# **Hierarchical Inheritance**

```python
class Calculator:
#     a=10
#     b=3
    def add(self,a,b):
        print("Sum is:",a+b)
    def mul(self,a,b):
        print("Multiplicatin is: ",a*b)
    def mod(self,a,b):
        print("Modulus is: ",a%b)
        
```

```python
class ADD(Calculator):
    def print_(self):
        print("This is from ADD SUb class")
```

```python
class Multi(Calculator):
    def print_(self):
        print("This is from Multiplicatin SUb class")
```

```python
class MOD(Calculator):
    def print_(self):
        print("This is from Modulus SUb class")
```

```python
m=int(input("Enter 1st number: "))
n=int(input("Enter 2nd number: "))
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
obj1=ADD()
obj1.print_()
obj1.add(m,n)
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
obj2=Multi()
obj2.print_()
obj2.mul(m,n)
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
obj3=MOD()
obj3.print_()
obj3.mod(m,n)
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

### **Hierarchical Inheritance End**

# **Hybrid Inheritance**

```python
class Addition:
    def add(self,a,b,c):
        print("Total marks is: ",a+b+c)
```

```python
class Avg:
    def avg(self,a,b,c):
        print("Average marks is: ",(a+b+c/3))
        
```

```python
class Cal(Addition,Avg):
    def _print(self):
        print("This is from Calculation class.")
```

```python
class Mark(Cal):
    def print_marks(self):
        print("Marks are-------- ")
```

```python
mark1=int(input("Enter 1st marks: "))
mark2=int(input("Enter 2nd marks: "))
mark3=int(input("Enter 3rd marks: "))
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
obj=Mark()
obj._print()
obj.print_marks()
obj.add(mark1,mark2,mark3)
obj.avg(mark1,mark2,mark3)
```

### **Hybrid Inderitance End**

# **Method Overriden**

```python
class overrriden:
    def overriden_(self):
        print("Hi I am Define in Base Class")
    def fun(self):
        print("This is a normal function")
class derived_1(overrriden):
    def fun(self):
        print("This is a normal function in Derived-1 class")
class derived_2(overrriden):
    def fun(self):
        print("This is a normal function in Derived-2 class") 
obj_overriden=overrriden()
obj_derived_1=derived_1()
obj_derived_2=derived_2()

for obj in (obj_overriden,obj_derived_1,obj_derived_2):
    print()
    obj.overriden_()
    obj.fun()
```

### **Method Overriden End**

# **Polymorphism**

```python
class A:
    def _print(self):
        print("This is A")
class B:
    def _print(self):
        print("This is B")
class C:
    def _print(self):
        print("This is C")
        
obj_A=A()  
obj_B=B()        
obj_C=C()

for obj in (obj_A,obj_B,obj_C):
    obj._print()
```

### **Ploymorphism End**
